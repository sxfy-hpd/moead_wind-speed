```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import mpl_toolkits.axisartist as axisartist
```

```python
df = pd.read_csv("../result/washingtong_spring_20120401-20120407新/three/first_pop.csv",header=None)
df_pareto = pd.read_csv("../result/washingtong_spring_20120401-20120407新/three/pareto.csv",header=None)
x1 = df.iloc[:,5]
y1 = df.iloc[:,6] * 10
x2 = df_pareto.iloc[:,5]
y2 = (df_pareto.iloc[:,6] *10)
```

```python
fig1, ax1 = plt.subplots(figsize=(4, 2.7), layout='constrained')
plt.ylim((0.62,0.74))
# plt.xlim((02,1.02))
plt.yticks(fontproperties = 'Times New Roman', size = 10)
plt.xticks(fontproperties = 'Times New Roman', size = 10)
# Hide the right and top spines
ax1.spines.right.set_visible(False)
ax1.spines.top.set_visible(False)
ax1.scatter(x1, y1,marker='+',color='y',label='Initial population')
ax1.scatter(x2, y2,marker='*',color='blue',label='non-dominated solution')
ax1.set_xlabel('f1',fontdict={'family' : 'Times New Roman', 'size'   : 10})  # Add an x-label to the axes.
ax1.set_ylabel(r'f2 × $10^{-1}$',fontdict={'family' : 'Times New Roman', 'size'   : 10})  # Add a y-label to the axes.
ax1.annotate('select', xy=(0.27,0.718), xytext=(0.35, 0.734),
            arrowprops=dict(facecolor='yellow', shrink=0.05,))
ax1.legend(prop={'family' : 'Times New Roman', 'size'   : 5}) # Add a legend.
ax1.grid(True,alpha=0.2)
#bbox_inches='tight'表示指定将图表多余的空白区域裁减掉
plt.savefig('washingtong_spring.pop.png',dpi=600,bbox_inches='tight')
plt.show()
```

```python
df = pd.read_csv("../result/washingtong_summer_20120701-20120707新/five/first_pop.csv",header=None)
df_pareto = pd.read_csv("../result/washingtong_summer_20120701-20120707新/five/pareto.csv",header=None)
x1 = df.iloc[:,5] + 0.1
y1 = df.iloc[:,6] * 10
x2 = df_pareto.iloc[:,5]
y2 = df_pareto.iloc[:,6] *10
```

```python
x11 = []
for x in x1:
    if x>1.0:
        x = 1.0
    x11.append(x)
```

```python
fig1, ax1 = plt.subplots(figsize=(4, 2.7), layout='constrained')
plt.ylim((0.575,0.70))
# plt.xlim((0.3,1.02))
plt.yticks(fontproperties = 'Times New Roman', size = 10)
plt.xticks(fontproperties = 'Times New Roman', size = 10)
# Hide the right and top spines
ax1.spines.right.set_visible(False)
ax1.spines.top.set_visible(False)
ax1.scatter(x11, y1,marker='+',color='y',label='Initial population')
ax1.scatter(x2, y2,marker='*',color='blue',label='non-dominated solution')
ax1.set_xlabel('f1',fontdict={'family' : 'Times New Roman', 'size'   : 10})  # Add an x-label to the axes.
ax1.set_ylabel(r'f2 × $10^{-1}$',fontdict={'family' : 'Times New Roman', 'size'   : 10})  # Add a y-label to the axes.
ax1.annotate('select', xy=(0.255,0.665), xytext=(0.32, 0.69),
            arrowprops=dict(facecolor='yellow', shrink=0.05))
ax1.legend(prop={'family' : 'Times New Roman', 'size'   : 5}) # Add a legend.
ax1.grid(True,alpha=0.2)
#bbox_inches='tight'表示指定将图表多余的空白区域裁减掉
plt.savefig('washingtong_winter_pop.png',dpi=600,bbox_inches='tight')
plt.show()
```
