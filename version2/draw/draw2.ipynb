```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def create_image(true,down,up,filename='California_summer.png',label='proposed'):
    t = np.arange(len(true))
    fig, ax = plt.subplots(figsize=(3.5, 2.7), layout='constrained')
    plt.ylim((8,15))
    plt.yticks(fontproperties = 'Times New Roman', size = 10)
    plt.xticks(fontproperties = 'Times New Roman', size = 10)
    ax.plot(t, true, lw=1, label='true',color='black')
    ax.fill_between(t, up, down, facecolor='C012', alpha=0.4,label=label)
    ax.legend(loc='upper right',prop={'family' : 'Times New Roman', 'size'   : 9.5})
    ax.set_xlabel('Samples(10min/point)',fontdict={'family' : 'Times New Roman', 'size'   : 10}) 
    ax.set_ylabel('Wind speed(m/s)',fontdict={'family' : 'Times New Roman', 'size'   : 10}) 
    ax.grid(True,alpha=0.4)
    plt.savefig(filename,dpi=600,bbox_inches='tight')
    plt.show()
```

```python
df_1 = pd.read_csv('../result/washingtong_autumn_20121001-20121007æ–°/five/picp:0.98500-pinrw:0.21374-cwc:2.28244.csv',header=None)
down = df_1.iloc[:,0]
up = df_1.iloc[:,1]
true = df_1.iloc[:,2]
```

```python
create_image(true,down,up,filename='Tang_predict.png',label='Tang\'s model')
```
